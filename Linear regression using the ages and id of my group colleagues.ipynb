```python
import csv
```

```python
# define the path to your desktop and the filename of the CSV file
test_age = "D:/Escritorio/test_age.csv"
```

```python
# write data to CSV file
with open(test_age, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for row in data:
        writer.writerow(row)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data from CSV file
datos = pd.read_csv("D:/Escritorio/test_age.csv")
```

```python
# check column names in the DataFrame
print(datos.columns)
```

```python
# extract the columns of interest
tiempo = datos["id"]
precio = datos["age"]
```

```python
# plot the data
plt.plot(tiempo, precio)
plt.xlabel("id (seq)")
plt.ylabel("age (seq)")
plt.show()
```

```python
# trazar el grfico de dispersin
plt.scatter(tiempo, precio)
plt.xlabel("id (seq)")
plt.ylabel("age (seq)")
plt.title("Predicted age")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
X = datos[["id", "age"]]
```

```python
# perform KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
# add the cluster labels to the DataFrame
datos["Cluster"] = kmeans.labels_
```

```python
# visualize the data using color-coded clusters
plt.scatter(x=datos["id"], y=datos["age"], c=datos["Cluster"])
plt.xlabel("id (seq)")
plt.ylabel("age (age)")
plt.show()
```
